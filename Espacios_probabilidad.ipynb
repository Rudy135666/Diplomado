# Axiomas de Probabilidad seg√∫n Kolmogorov

La teor√≠a moderna de la probabilidad se fundamenta en los **axiomas de Kolmogorov (1933)**, que definen formalmente un **espacio de probabilidad** $((\Omega, \mathcal{F}, P))$, donde:

- $(\Omega)$ es el **espacio muestral**: el conjunto de todos los posibles resultados de un experimento aleatorio.  
- $(\m$athcal{F})$ es una **œÉ-√°lgebra** de subconjuntos de $(\Omega)$: un conjunto de eventos sobre los que se define la probabilidad.  
- $(P)$ es una **funci√≥n de probabilidad**: asigna a cada evento $(A \in \mathcal{F})$ un n√∫mero $(P(A)\in [0,1]).$

## Axiomas de Kolmogorov

1. **Axioma de no negatividad:**

$$
P(A) \ge 0, \quad \forall A \in \mathcal{F}
$$

La probabilidad de cualquier evento es un n√∫mero no negativo.

2. **Axioma de normalizaci√≥n:**

$$
P(\Omega) = 1
$$

El evento seguro (que ocurre siempre) tiene probabilidad 1.

3. **Axioma de aditividad (o œÉ-aditividad):**

Si $(A_1, A_2, A_3, \dots)$ son eventos **mutuamente excluyentes** (disjuntos dos a dos), entonces:

$$
P\left(\bigcup_{i=1}^{\infty} A_i \right) = \sum_{i=1}^{\infty} P(A_i)
$$

La probabilidad de la uni√≥n de eventos disjuntos es igual a la suma de sus probabilidades individuales.

## Consecuencias importantes

A partir de estos axiomas se derivan propiedades fundamentales de la probabilidad, como:

- Probabilidad del complemento: $(P(A^c) = 1 - P(A))$  
- Monoton√≠a: si $(A \subseteq B)$, entonces $(P(A) \le P(B))$  
- Probabilidad del conjunto vac√≠o: $(P(\emptyset) = 0)$  
- F√≥rmula de inclusi√≥n-exclusi√≥n para dos o m√°s eventos

**Nota:**  
Estos axiomas proporcionan una base **rigorosa y general** para cualquier desarrollo de la teor√≠a de probabilidad, incluyendo variables aleatorias, distribuciones y procesos estoc√°sticos.


## 1. Enfoque cl√°sico de la probabilidad
En este enfoque, la probabilidad se define como la raz√≥n entre los casos favorables y los casos posibles.

### **Definiciones:**
- **Espacio muestral ($\Omega$):** En la teor√≠a de probabilidades, el espacio muestral o espacio de muestreo consiste en el conjunto de todos los posibles resultados de un experimento aleatorio, junto con una estructura sobre el mismo.

- **Evento:** En probabilidad, un evento (o suceso) es un conjunto de uno o m√°s resultados posibles de un experimento aleatorio. Es un subconjunto del espacio muestral, que es la colecci√≥n de todos los resultados posibles de dicho experimento.

Para un experimento aleatorio, se tiene $\Omega$ es el espacio muestral.

Sea $A$ un evento, entonces:
$$
P(A) = \frac{|A|}{|\Omega|}
$$
**Ejemplo:** Sea el experimento aleatoria de mandar una moneda 3 veces, de manera independiente, es decir, el $i-√©simo$ lanzamiento no me afecta al $i+1$ lanzamiento. El espacio muestral de este experimento es:

$$\Omega = \{A,S\} \times \{A,S\} \times \{A,S\} \times = \{A,S\}^3$$
y se tiene que $|\Omega| = 2^{3}$.

-$A$: Todos los resultados son √°guila

### **Probabilidad Condicional**

El estudio de la probabilidad tiene sus or√≠genes en el **siglo XVII**, cuando los matem√°ticos **Blaise Pascal** y **Pierre de Fermat** intercambiaron cartas para resolver problemas de **juegos de azar**.  
Posteriormente, **Jacob Bernoulli** formul√≥ la **Ley de los Grandes N√∫meros (1713)**, y en el siglo XIX **Pierre-Simon Laplace** expandi√≥ el an√°lisis probabil√≠stico.  

Finalmente, en 1933, **Andrey Kolmog√≥rov** formaliz√≥ el enfoque **axiom√°tico de la probabilidad**, que hoy en d√≠a es la base te√≥rica moderna.  

La probabilidad condicional en estad√≠stica mide la probabilidad de que un determinado evento ocurra en funci√≥n de la ocurrencia (o no) de otros eventos relacionados. Tiene diferentes aplicaciones en ciencia y finanzas.

En el estudio de la probabilidad condicional, los investigadores examinan dos o m√°s eventos con probabilidades relacionadas y se preguntan: ¬´Si sabemos que A ocurri√≥, ¬øcu√°l es la probabilidad de que B tambi√©n ocurra?¬ª. La probabilidad se calcula multiplicando la probabilidad del evento anterior por la probabilidad actualizada del evento posterior, o condicional. Tambi√©n es posible calcular la probabilidad de que un evento anterior ocurriera, bas√°ndose en la ocurrencia (conocida) del evento posterior.

**Definimos** la probabilidad de que ocurra un evento $A$ **sabiendo** que ocurre un evento $B$,
$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}, \quad P(B) > 0
$$
donde $\mathbb{P}(B)>0$.




```python
#Vamos a definir la funci√≥n de la probabilidad condicional

def prob_cond(E,F):
    return len(E & F)/ len(F)
```

#### üé≤ Problema de la urna: 5 rojas y 7 azules

Consideremos una **urna con 5 bolas rojas y 7 bolas azules** (en total 12 bolas).  
Estudiaremos distintas probabilidades de extracci√≥n, con y sin reemplazo,  
utilizando **definiciones cl√°sicas**, **probabilidad condicional**,  
y la **distribuci√≥n hipergeom√©trica**.

Definiciones iniciales:

- \(N = 12\) (total de bolas).
- \(K = 5\) (n√∫mero de bolas rojas).
- \(N-K = 7\) (n√∫mero de bolas azules).

Trabajaremos en los siguientes escenarios:
1. **Una sola extracci√≥n.**
2. **Dos extracciones sin reemplazo.**
3. **Dos extracciones con reemplazo.**
4.  **Probabilidades condicionales.**
5. **Verificaci√≥n por simulaci√≥n en Python.**


##### 1Ô∏è‚É£ Una extracci√≥n

El espacio muestral contiene 12 resultados equiprobables.  

- Probabilidad de extraer una roja:
$$
P(\text{roja}) = \frac{5}{12}.$$

- Probabilidad de extraer una azul:
$$
P(\text{azul}) = \frac{7}{12}.$$


##### 2Ô∏è‚É£ Dos extracciones sin reemplazo

a) Dos rojas
$$
P(\text{2 rojas}) = \frac{\binom{5}{2}}{\binom{12}{2}} = \frac{10}{66} = \frac{5}{33}.
$$

b) Roja y luego azul
$$
P(\text{roja luego azul}) = \frac{5}{12}\cdot\frac{7}{11} = \frac{35}{132}.
$$

c) Al menos una roja
$$
P(\text{al menos una roja}) = 1 - P(\text{2 azules}).
$$
$$
P(\text{2 azules}) = \frac{\binom{7}{2}}{\binom{12}{2}} = \frac{21}{66} = \frac{7}{22}.
$$
Por lo tanto:
$$
P(\text{al menos una roja}) = 1 - \frac{7}{22} = \frac{15}{22}.
$$


#### 3Ô∏è‚É£ Dos extracciones con reemplazo

Si despu√©s de cada extracci√≥n devolvemos la bola y mezclamos, las extracciones son **independientes**:

$$
P(\text{2 rojas}) = \left(\frac{5}{12}\right)^2 = \frac{25}{144}.
$$


##### 4Ô∏è‚É£ Probabilidades condicionales

- $$(P(\text{2¬™ roja} \mid \text{1¬™ roja}) = \frac{4}{11}).$$
- $$(P(\text{2¬™ roja} \mid \text{1¬™ azul}) = \frac{5}{11}).$$
- $$(P(\text{2¬™ roja}) = \frac{5}{12}).$$

Conclusi√≥n: **no hay independencia** entre extracciones sin reemplazo.


```python
import random
import numpy as np
from math import comb
import matplotlib.pyplot as plt

# Par√°metros
N, K = 12, 5
urna = ["R"]*K + ["A"]*(N-K)

# ----- Simulaci√≥n sin reemplazo -----
n_sim = 200_000
count_rr = count_ra = count_ar = count_aa = 0

for _ in range(n_sim):
    muestra = random.sample(urna, 2)  # sin reemplazo
    if muestra[0]=="R" and muestra[1]=="R":
        count_rr += 1
    elif muestra[0]=="R" and muestra[1]=="A":
        count_ra += 1
    elif muestra[0]=="A" and muestra[1]=="R":
        count_ar += 1
    else:
        count_aa += 1

print("üîπ Simulaci√≥n SIN reemplazo (ordenado):")
print(f"P(RR) sim: {count_rr/n_sim:.3f}")
print(f"P(RA) sim: {count_ra/n_sim:.3f}")
print(f"P(AR) sim: {count_ar/n_sim:.3f}")
print(f"P(AA) sim: {count_aa/n_sim:.3f}")

# ----- Teor√≠a hipergeom√©trica -----
p_2rojas = comb(5,2)/comb(12,2)
p_1roja = (comb(5,1)*comb(7,1))/comb(12,2)
p_0rojas = comb(7,2)/comb(12,2)

print("\nüîπ Teor√≠a sin reemplazo:")
print(f"P(2 rojas): {p_2rojas:.3f}")
print(f"P(1 roja): {p_1roja:.3f}")
print(f"P(0 rojas): {p_0rojas:.3f}")

# ----- Simulaci√≥n con reemplazo -----
count_rr = count_ra = count_ar = count_aa = 0
for _ in range(n_sim):
    x = random.choice(urna)
    y = random.choice(urna)
    if x=="R" and y=="R":
        count_rr += 1
    elif x=="R" and y=="A":
        count_ra += 1
    elif x=="A" and y=="R":
        count_ar += 1
    else:
        count_aa += 1

print("\nüîπ Simulaci√≥n CON reemplazo (ordenado):")
print(f"P(RR) sim: {count_rr/n_sim:.3f}")
print(f"P(RA) sim: {count_ra/n_sim:.3f}")
print(f"P(AR) sim: {count_ar/n_sim:.3f}")
print(f"P(AA) sim: {count_aa/n_sim:.3f}")
print(f"P(2 rojas) con reemplazo te√≥rica: {(K/N)**2:.3f}")

# ----- Convergencia frecuencia relativa -----
M = 20000
res = [random.choice(urna) for _ in range(M)]
acc = np.cumsum([1 if r=="R" else 0 for r in res])
freq_rel = acc / np.arange(1, M+1)

plt.figure(figsize=(8,3.5))
plt.plot(freq_rel, label="Frecuencia relativa (R)")
plt.axhline(K/N, color="red", linestyle="--", label=f"Te√≥rico = {K}/{N}={K/N:.3f}")
plt.xlabel("N√∫mero de extracciones")
plt.ylabel("Frecuencia relativa de R")
plt.legend()
plt.title("Convergencia de frecuencia relativa hacia la probabilidad te√≥rica")
plt.show()
```

### **Independencia de eventos**

La independencia de eventos en probabilidad ocurre cuando la ocurrencia de un evento no afecta la probabilidad de que ocurra otro evento. Matem√°ticamente, dos eventos A y B son independientes si la probabilidad de que ocurran ambos (la intersecci√≥n) es igual al producto de sus probabilidades individuales:

Decimos que dos eventos $A$ y $B$ son independientes si
$$\mathbb{P}(A \cap B) = \mathbb{P}(A)\mathbb{P}(B)$$
O de manera equivalente
$$\mathbb{P}(A|B) \mathbb{P}(A)$$

```python
def indep(E,F): #definimos la independencia de E y F
    return prob(E & F) == prob(E)*prob(F)  
```

## 2. Enfoque geom√©trico de la probabilidad
El enfoque geom√©trico en probabilidad representa la probabilidad de un evento como una proporci√≥n de una medida (longitud, √°rea o volumen) de un espacio muestral geom√©trico. Es √∫til para variables aleatorias continuas, calculando la probabilidad de que un evento ocurra dentro de un objeto geom√©trico al dividir el "espacio favorable" entre el "espacio total.

Este enfoque se aplica cuando los resultados corresponden a puntos en una regi√≥n geom√©trica.
La probabilidad se calcula como:
$$
P(A) = \frac{(A)}{(\Omega)}
$$

**Ejemplo:**

Dos personas llegan al azar en un intervalo de una hora. Cada una espera 15 minutos. La probabilidad de que se encuentren corresponde al √°rea de una regi√≥n en el cuadrado unidad.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000
X = np.random.rand(N)
Y = np.random.rand(N)

# Condici√≥n de encuentro: |X - Y| <= 0.25
mask = np.abs(X - Y) <= 0.25
P_encuentro = np.mean(mask)

plt.figure(figsize=(5,5))
plt.scatter(X[~mask], Y[~mask], alpha=0.2, s=2, label="No se encuentran")
plt.scatter(X[mask], Y[mask], alpha=0.2, s=2, label="Se encuentran")
plt.xlabel("Tiempo llegada persona A")
plt.ylabel("Tiempo llegada persona B")
plt.legend()
plt.title("Simulaci√≥n del problema del encuentro")
plt.show()

P_encuentro
```

**Ejemplo:**

Romero y Julieta planean encontrarse en un lugar especifico entre [0,60]
- Si Romero llega primero, entonces piensa esperar 10 min.
- Si Julieta llega primero, entonces piensa esperar 10 min.

Se elige un n√∫mero al azar entre 0 y 60, supongamos que al par de n√∫meros se elige sobre el cuadro de modo que la probabilidad de un evento sea el
√°rea favorable entre el √°rea total.

Sea el evento
A: Romeo y Julieta se encuentran a una distancia de 10 min.

```python
import numpy as np
import matplotlib.pyplot as plt

#Vamos a generar 100,000 n√∫meros aleatorios
muestra=1000000

#Generamos los tiempos aleatorios entre [0,60]
#x : tiempo de llegada de Romero
#y : tiempo de llegada de Julieta
x=np.random.uniform(0,60,muestra) #genera 100,000 n√∫meros aleatorios entre 0 y 60  
y=np.random.uniform(0,60,muestra) #genera 100,000 n√∫meros aleatorios entre 0 y 60

#vamos a a√±adir una condici√≥n para que se encuentren
encuentro = np.where(abs(x-y)<=10, 'r', 'b') 
#si la diferencia entre x y y es menor o igual a 10, entonces se encuentran (color rojo), si no, no se encuentran (color azul)

#Vamos a calcular la probabilidad de que se encuentren
k=0
for element in encuentro:
    if element =='r':
        k+=1
print("La probabilidad del evento A es: {}.".format(k/muestra))


#Para gr√°ficar los puntos de encuentro, hacemos una gr√°fica de dispersi√≥n
plt.figure(figsize=(8,8))
plt.scatter(x,y,c=encuentro, s = 1)
plt.show()
```

## 3. Enfoque frecuentista de la probabilidad

El enfoque frecuentista de la probabilidad define la probabilidad de un evento como la frecuencia relativa con la que ocurre dicho evento en un gran n√∫mero de ensayos o repeticiones de un fen√≥meno aleatorio. Es decir, la probabilidad se estima como la proporci√≥n de veces que un suceso A ha ocurrido (F) sobre el n√∫mero total de ensayos (N), expresada como P(A) = F/N. Este enfoque se basa en la ley de los grandes n√∫meros, que establece que al aumentar el n√∫mero de observaciones, la frecuencia relativa se aproxima al l√≠mite de la probabilidad real. 

$$
P(A) \approx \lim_{n \to \infty} \frac{N(A)}{n}
$$

Ejemplo cl√°sico: lanzamiento de una moneda (volado).

```python

import random #Libreria para generar n√∫meros aleatorios

def volado(n):
    resultados = [random.choice(["√Åguila", "Sol"]) for _ in range(n)]
    freq_aguila = resultados.count("√Åguila") / n
    freq_sello = resultados.count("Sol") / n
    return freq_aguila, freq_sello

for n in [10, 100, 1000, 10000]: # n va a ser el n√∫mero de lanzamientos 
    print(f"n={n}: {volado(n)}")
```
